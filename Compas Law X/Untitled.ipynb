```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def draw_system_architecture():
    fig, ax = plt.subplots(figsize=(9, 6))

    ax.set_xlim(0, 12)
    ax.set_ylim(0, 12)

    # Draw the layers
    application_layer = mpatches.FancyBboxPatch((2, 9), 8, 2, boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightgrey')
    consensus_layer = mpatches.FancyBboxPatch((2, 6), 8, 2, boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightgrey')
    network_layer = mpatches.FancyBboxPatch((2, 3), 8, 2, boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightgrey')
    data_layer = mpatches.FancyBboxPatch((2, 0), 8, 2, boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightgrey')

    ax.add_patch(application_layer)
    ax.add_patch(consensus_layer)
    ax.add_patch(network_layer)
    ax.add_patch(data_layer)

    # Add text to layers
    ax.text(6, 10, 'Application Layer\n(User Interfaces, Applications)', ha='center', va='center', fontsize=12)
    ax.text(6, 7, 'Consensus Layer\n(Mechanisms, Proof of Stake)', ha='center', va='center', fontsize=12)
    ax.text(6, 4, 'Network Layer\n(Network of Nodes)', ha='center', va='center', fontsize=12)
    ax.text(6, 1, 'Data Layer\n(Blockchain Ledger, Database)', ha='center', va='center', fontsize=12)

    # Add labels
    ax.text(0, 10, '1', ha='center', va='center', fontsize=12, color='blue')
    ax.text(0, 7, '2', ha='center', va='center', fontsize=12, color='blue')
    ax.text(0, 4, '3', ha='center', va='center', fontsize=12, color='blue')
    ax.text(0, 1, '4', ha='center', va='center', fontsize=12, color='blue')

    ax.axis('off')
    plt.title('System Architecture Diagram', fontsize=14)
    plt.savefig('system_architecture.png', dpi=300)  # dpi controls the resolution
    plt.show()

draw_system_architecture()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_land_registration_workflow():
    G = nx.DiGraph()

    # Add nodes and edges
    nodes = [
        "User Submission", "Data Entry", "Smart Contract Activation",
        "Verification", "Consensus Mechanism", "Blockchain Update", "Confirmation"
    ]
    edges = [
        ("User Submission", "Data Entry"),
        ("Data Entry", "Smart Contract Activation"),
        ("Smart Contract Activation", "Verification"),
        ("Verification", "Consensus Mechanism"),
        ("Consensus Mechanism", "Blockchain Update"),
        ("Blockchain Update", "Confirmation")
    ]

    G.add_nodes_from(nodes)
    G.add_edges_from(edges)

    pos = {
        "User Submission": (0, 5), "Data Entry": (1, 4), "Smart Contract Activation": (2, 3),
        "Verification": (3, 2), "Consensus Mechanism": (4, 1), "Blockchain Update": (5, 2), "Confirmation": (6, 3)
    }

    plt.figure(figsize=(12, 6))
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True)
    plt.title('Land Registration Workflow')
    plt.show()

draw_land_registration_workflow()
```

```python
import matplotlib.pyplot as plt

def draw_implementation_roadmap():
    fig, ax = plt.subplots(figsize=(12, 6))

    phases = ['Pilot Phase', 'Gradual Rollout', 'Full Implementation']
    durations = [1, 3, 5]  # Representing years for each phase

    bars = plt.barh(phases, durations, color='skyblue')

    for bar in bars:
        plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f'{bar.get_width()} years', ha='center', va='center')

    plt.xlabel('Duration (Years)')
    plt.title('Implementation Roadmap')
    plt.xlim(0, 6)
    plt.show()

draw_implementation_roadmap()
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='System Architecture Diagram')

# Add nodes for different sections
dot.node('Users', 'Users\nWeb Client | Mobile Client | Admin Dashboard', shape='box')
dot.node('API Gateway', 'API Gateway', shape='box')
dot.node('Microservices', '''Microservices
- User Management
- Property Management
- Payments
- Transaction Management
- Notification
- Dispute Resolution
- Report Generation
- GIS Integration''', shape='box')
dot.node('Integration Layer', '''Integration Layer
- Government DBs
- Payment Gateways
- GIS System''', shape='box')
dot.node('Blockchain Network', '''Blockchain Network
- Validator Nodes
- Smart Contracts
- Consensus
- IPFS
- Off-chain DB
- Node Network''', shape='box')
dot.node('Cloud Infrastructure', '''Cloud Infrastructure & Security
- CI/CD Pipeline
- Monitoring & Logs
- Backup
- Security Services
- Load Balancers
- Storage''', shape='box')

# Add edges to show connections
dot.edge('Users', 'API Gateway')
dot.edge('API Gateway', 'Microservices')
dot.edge('Microservices', 'Integration Layer')
dot.edge('Integration Layer', 'Blockchain Network')
dot.edge('Blockchain Network', 'Cloud Infrastructure')

# Save and render the graph
dot.render('system_architecture_diagram', format='png')
```
